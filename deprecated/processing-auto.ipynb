```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib as plt
```

```python
scope_file = "scope_readout/nbiot_test1_coap_psm5min1.csv"

#Getting the amount of points in each segment
df = pd.read_csv(scope_file, header = [0,1])

mean = df.iloc[1:500].mean()
std = df.iloc[1:500].std()

filtered = df.where(df > (mean+2*std))
filtered.plot()
```
